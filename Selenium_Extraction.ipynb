```python
from bs4 import BeautifulSoup 
from selenium import webdriver 
from time import sleep
import pandas as pd
import numpy as np
import csv

def transform_html(input):
    return " ".join(input.split()).replace('> <', '><')
```

```python
# some definitions and general informations about the matches


id_match = 1845
first_statement = 1559783
last_statement = 1559784
championship = 'Premier League'
edition = '2021-2022'
country = 'England'



# create of csv file where the informations will be save 

columns = 'id_match,names,home_away,managers,formations,attendance,date,stadium,championship,edition,country,score,goals,goals_past,total_shots,woodwork,shots_on_target,shots_off_target,blocked_shots,total_possession,touches,total_passes,accurate_passes,pass_success,key_passes,dribbles_attempted,dibles,dribbles_past,dribbles_success,aerials_won,aerial_success,offensive_aerials,defensive_aerials,tackles,tackle_success,tackles_attempted,was_dribbled,clearances,interceptions,corners,corner_accuracy,dispossessed,errors,fouls,offsides,start_minutes,end_minutes'
with open ('C:/Users/lesll/Downloads/archive/premier_league.csv', 'a') as archive:
    archive.write(columns)

#range number is interval of pages(url) on site 1xbet about a specific championship
for url in range (first_statement, last_statement+1):
    
    # open browser and waiting 30 seconds to site dont realize thats not human
    browser = webdriver.Chrome()
    browser.get(f'https://1xbet.whoscored.com/Matches/{url}/Live')
    sleep (30)
    
    # after the robot entered on the site, he take the html
    # thats important because sometimes sites use the react and if you try do this only with
    # BeautifulSoup and Requests (or urllib), its gonna be error
    
    html = browser.page_source
    html = transform_html(html)
    page = BeautifulSoup(html, 'html.parser')
    
    # the function find and findAll help us to browsing into the page with specific searches 
    stats = page.findAll('div', class_ = "match-centre-stats")[0]
    
    # here with filter all important informations about the statistics and match specifics
    # like team names, stadium, date and more
    info_teams = page.findAll('div', class_ = "team-info")
    home_team = info_teams[0]
    away_team = info_teams[1]
    home_name = home_team.find('a', {'class': "team-name"}).get_text()
    away_name = away_team.find('a', {'class': "team-name"}).get_text()
    home_manager = home_team.find('div', {'class': "manager"}).get_text()
    away_manager = away_team.find('div', {'class': "manager"}).get_text()
    home_formation = home_team.find('div', {'class': "formation"}).get_text()
    away_formation = away_team.find('div', {'class': "formation"}).get_text()

    if len(page.findAll('span', class_ = "attendance")) != 0:
        attendance = page.findAll('span', class_ = "attendance")[0].get_text()
    if len(page.findAll('span', class_ = "attendance")) == 0:
        attendance = 0

    date = page.findAll('dd')[4].get_text()
    stadium = page.findAll('span', class_ ="venue")[0].get_text()
    score = page.find('div', {'class' : 'score'}).get_text().replace(':', '-')
    goals = [score[0], score[-1]]
    goals_past = [score[-1], score[0]]
    
    names = [home_name, away_name]
    managers = [home_manager, away_manager]
    formations = [home_formation, away_formation]
    
    # This part will take the statistics so specifics (Thats site is so great!!)
    
    # total of shots
    place_shots = stats.findAll('li')[1]
    total_shots = []
    place_shots.findAll('span')
    for i in place_shots.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            total_shots.append(int(a))

    # woodwork 
    place_woodwork = stats.findAll('li')[12]
    woodwork = []
    for i in place_woodwork.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            woodwork.append(int(a))        

    # shots on target
    place_shots_on_target = stats.findAll('li')[13]
    shots_on_target = []
    for i in place_shots_on_target.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            shots_on_target.append(int(a))        

    # shots off target
    place_shots_off_target = stats.findAll('li')[14]
    shots_off_target = []
    for i in place_shots_off_target.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            shots_off_target.append(int(a))        

     # blocked shots
    place_blocked_shots = stats.findAll('li')[15]
    blocked_shots = []
    for i in place_blocked_shots.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            blocked_shots.append(int(a))

    # possession
    place_possession = stats.findAll('li')[2]
    total_possession = []
    for i in place_possession.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            total_possession.append(float(a))

    # touches
    place_touches = stats.findAll('li')[18]
    touches = []
    for i in place_touches.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            touches.append(int(a))

    # total passes 
    place_total_passes = stats.findAll('li')[21]
    total_passes = []
    for i in place_total_passes.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            total_passes.append(int(a))

    # accurate passes
    place_accurate_passes = stats.findAll('li')[22]
    accurate_passes = []
    for i in place_accurate_passes.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            accurate_passes.append(int(a))

    # pass success
    place_pass_success = stats.findAll('li')[3]
    pass_success = []
    for i in place_pass_success.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            pass_success.append(int(a))

    # key passes
    place_key_passes = stats.findAll('li')[23]
    key_passes = []
    for i in place_key_passes.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            key_passes.append(int(a))

    # dribbles attempted 
    place_dribbles_attempted = stats.findAll('li')[26]
    dribbles_attempted = []
    for i in place_dribbles_attempted.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            dribbles_attempted.append(int(a))

    # dibles
    place_dibles = stats.findAll('li')[4]
    dibles = []
    for i in place_dibles.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            dibles.append(int(a))

    # dribbles past
    place_dribbles_past = stats.findAll('li')[27]
    dribbles_past = []
    for i in place_dribbles_past.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            dribbles_past.append(int(a))

    # dribbles success
    place_dribbles_success = stats.findAll('li')[28]
    dribbles_success = []
    for i in place_dribbles_success.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            dribbles_success.append(int(a))

    # aerials won
    place_aerials_won = stats.findAll('li')[5]
    aerials_won = []
    for i in place_aerials_won.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            aerials_won.append(int(a))

    # Aerials Success
    place_aerial_success = stats.findAll('li')[31]
    aerial_success = []
    for i in place_aerial_success.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            aerial_success.append(int(a))

    # Offensive Aerials
    place_offensive_aerials = stats.findAll('li')[32]
    offensive_aerials = []
    for i in place_offensive_aerials.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            offensive_aerials.append(int(a))

    # defensive aerials 
    place_defensive_aerials = stats.findAll('li')[33]
    defensive_aerials = []
    for i in place_defensive_aerials.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            defensive_aerials.append(int(a))

    # tackles
    place_tackles = stats.findAll('li')[6]
    tackles = []
    for i in place_tackles.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            tackles.append(int(a))

    # Tackle Success 
    place_tackle_success = stats.findAll('li')[38]
    tackle_success = []
    for i in place_tackle_success.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            tackle_success.append(int(a))


    # Tackles Attempted 
    place_tackles_attempted = stats.findAll('li')[36]
    tackles_attempted = []
    for i in place_tackles_attempted.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            tackles_attempted.append(int(a))

    # Was Dribbled
    place_was_dribbled = stats.findAll('li')[37]
    was_dribbled = []
    for i in place_was_dribbled.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            was_dribbled.append(int(a))

    # Clearances 
    place_clearances = stats.findAll('li')[39]
    clearances = []
    for i in place_clearances.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            clearances.append(int(a))

    # Interceptions
    place_interceptions = stats.findAll('li')[40]
    interceptions = []
    for i in place_interceptions.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            interceptions.append(int(a))

    # corners
    place_corners = stats.findAll('li')[7]
    corners = []
    for i in place_corners.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            corners.append(int(a))

    # corner Accuracy
    place_corner_accuracy = stats.findAll('li')[43]
    corner_accuracy = []
    for i in place_corner_accuracy.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            corner_accuracy.append(int(a))

    # dispossessed 
    place_dispossessed = stats.findAll('li')[8]
    dispossessed = []
    for i in place_dispossessed.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            dispossessed.append(int(a))

    # errors
    place_errors = stats.findAll('li')[46]
    errors = []
    for i in place_errors.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            errors.append(int(a))

    # fouls
    place_fouls = stats.findAll('li')[47]
    fouls = []
    for i in place_fouls.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            fouls.append(int(a))

    # offsides
    place_offsides = stats.findAll('li')[48]
    offsides = []
    for i in place_offsides.findAll('span'):
        a = i.get_text()
        if a == '-' or a == '':
            pass
        else:

            offsides.append(int(a))
            
    # All statistics is saved in a list (just like that: shots_on_target[20,30])
    # the first index of this list is statistic about the home team and last index about the away team
    # so will be a loop with two statement, first statement will save all the informations of home team and after away team 
    # and then write this informations in a csv file
    # theres a column called home_away that identifies home team as 0 and away team as 1
    
    # home_away: 0 - home team______ 1 - away team
    
    for index_ in range(0,2):
        
        # the last columns theres 0 and 90 because in future projects I pretend make the extraction og this same data
        # but in different times of a match (just like half time)
        with open ('C:/Users/lesll/Downloads/archive/premier_league.csv', 'a') as archive:
            archive.write(f"\n{id_match},{names[index_]},{index_},{managers[index_]},{formations[index_]},{attendance},{date[5:]},{stadium},{championship},{edition},{country},{score},{goals[index_]},{goals_past[index_]},{total_shots[index_]},{woodwork[index_]},{shots_on_target[index_]},{shots_off_target[index_]},{blocked_shots[index_]},{total_possession[index_]},{touches[index_]},{total_passes[index_]},{accurate_passes[index_]},{pass_success[index_]},{key_passes[index_]},{dribbles_attempted[index_]},{dibles[index_]},{dribbles_past[index_]},{dribbles_success[index_]},{aerials_won[index_]},{aerial_success[index_]},{offensive_aerials[index_]},{defensive_aerials[index_]},{tackles[index_]},{tackle_success[index_]},{tackles_attempted[index_]},{was_dribbled[index_]},{clearances[index_]},{interceptions[index_]},{corners[index_]},{corner_accuracy[index_]},{dispossessed[index_]},{errors[index_]},{fouls[index_]},{offsides[index_]},0,90")
    
    # id_match identifes each match, so each statement of loop on url is about a different match 
    id_match += 1
    

    # The last part of code the csv file is save in a S3 Bucket on AWS
    # For it your computer need be connected with AWS account through AWS CLI
teste = pd.read_csv('C:/Users/lesll/Downloads/archive/premier_league1.csv', encoding='ANSI', index_col=False)
teste.to_csv('s3://nome_bucket/premier_league.csv', index = False)
        
```
